# **DATA CLEANING**

```python
# Replace incorrect values in 'sex' column
df['sex'].replace('.', pd.NA, inplace=True)

# Fill missing numerical values with median
num_cols = ["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing values in 'sex' with the most common value
df['sex'].fillna(df['sex'].mode()[0], inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Confirm missing values are handled
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Histogram for numerical variables
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Numerical Variables", fontsize=16)
plt.show()
```

## Numerical Variables (Histograms & Boxplots):

Culmen Length & Depth: The data is spread over a range, with some peaks indicating common values.
Flipper Length & Body Mass: These show normal-like distributions with some skewness.
Boxplots: Outliers are present, especially in body mass.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set theme for better aesthetics
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Create boxplot using seaborn with vibrant colors
sns.boxplot(data=df[["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]], palette="Set2")

# Enhance readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("Distribution of Numerical Features (Boxplot)", fontsize=16, fontweight='bold')
plt.xlabel("Feature Names", fontsize=14)
plt.ylabel("Values", fontsize=14)

# Show the plot
plt.show()
```

## Categorical Variables:

Species: Adelie and Gentoo are more frequent than Chinstrap.
Island: Biscoe has the highest count, followed by Dream and Torgersen.
Sex: The distribution of males and females is nearly equal, with slightly more males.

```python
# Count plots for categorical variables
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.countplot(x=df['species'], palette="viridis")
plt.title("Count of Species")

plt.subplot(1, 3, 2)
sns.countplot(x=df['island'], palette="coolwarm")
plt.title("Count of Islands")

plt.subplot(1, 3, 3)
sns.countplot(x=df['sex'], palette="pastel")
plt.title("Count of Sex")

plt.tight_layout()
plt.show()
```

```python
df['species'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title("Distribution of Penguin Species")
plt.ylabel("") # Remove default y-label
plt.show()
```

# BIVARIATE ANALYSIS

## 1. Scatter Plot (Numerical vs. Numerical)
Positive correlation likely between flipper length & body mass.
Different species may form distinct clusters (Adelie, Gentoo, Chinstrap).



```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x=df['flipper_length_mm'], y=df['body_mass_g'], hue=df['species'])
plt.title("Flipper Length vs. Body Mass by Species")
plt.show()
```

## 2. Box Plot (Categorical vs. Numerical)
Gentoo penguins might be heavier than Adelie & Chinstrap.
Spread (IQR) shows body mass variation within species.
Outliers could be detected in weight measurements.




```python
sns.boxplot(x=df['species'], y=df['body_mass_g'])
plt.title("Body Mass Distribution Across Species")
plt.show()
```

## 3. Heatmap (Correlation Matrix)
Strong positive correlation between body mass and flipper length.
Culmen length vs. depth might show a weak or negative correlation.
Low correlation means variables vary independently.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
